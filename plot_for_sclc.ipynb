```python
import matplotlib.pyplot as plt

# Study names
studies = ['CLCGP, Nat Genet 2012', 'Johns Hopkins, Nat Genet 2012', 'U Cologne, Nature 2015', 'Multi-Institute, Cancer Cell 2017']

# Total sample counts for each study
sample_counts = [29, 80, 120, 20]

# Creating the bar plot
plt.figure(figsize=(10, 6))
plt.bar(studies, sample_counts, color='skyblue')

plt.xlabel('Study')
plt.ylabel('Total Sample Count')
plt.title('Total Sample Counts in SCLC Studies')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['CLCGP, Nat Genet 2012', 'Johns Hopkins, Nat Genet 2012', 'U Cologne, Nature 2015', 'Multi-Institute, Cancer Cell 2017']
sizes = [29, 80, 120, 20]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = (0, 0.1, 0, 0)  # explode the 2nd slice (Johns Hopkins)

# Plot
plt.figure(figsize=(10, 7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart of Number of Samples in SCLC Studies')
plt.show()
```
